```python
import pandas as pa
import seaborn as sb
import matplotlib.pyplot as plt

df = pa.read_csv('/content/world_population.csv')
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.nunique()
```

```python
df.sort_values(by='World Population Percentage', ascending=False)
```

```python
df_corr = df.select_dtypes(include=['number'])
df_corr.corr()
```

```python
spearman = df_corr.corr()
spearman
```

```python
import matplotlib.pyplot as plt
import seaborn as sms
sms.heatmap(df_corr.corr(), annot=True,)
plt.rcParams["figure.figsize"] = (20,7)
plt.show()
```

```python
kendall = df_corr.corr('kendall')
kendall
```

```python
sb.heatmap(kendall, annot=True)
```
