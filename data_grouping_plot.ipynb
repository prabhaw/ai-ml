```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


df = pd.read_csv('/content/world_population.csv')
df.info()
```

```python
df1 = df[(df["Continent"] == "Oceania")]
df1
```

```python
df2 = df.groupby("Continent").mean(numeric_only=True).sort_values("2022 Population", ascending=False)
df2
```

```python
df3 = df2.transpose()
df3
```

```python
df4 = df3.plot()
df4
```

```python
df5 = df3.boxplot()
df5
```
